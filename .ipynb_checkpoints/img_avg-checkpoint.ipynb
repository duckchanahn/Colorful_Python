```python
# 진또배기

# image = cv2.imread(img_url)

def getImgAvg ( image ) :

    # 채널을 BGR -> RGB로 변경
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = image.reshape((image.shape[0] * image.shape[1], 3)) # height, width 통합

    count_color = 3
    clt = KMeans(n_clusters = count_color)
    clt.fit(image)

    colorList = []
    numLabels = np.arange(0, len(np.unique(clt.labels_)) + 1)
    (hist, _) = np.histogram(clt.labels_, bins=numLabels)

    # normalize the histogram, such that it sums to one
    hist = hist.astype("float")
    hist /= hist.sum()
    
    colorRatioList = list(hist)
    for (colorRatio, color) in zip(colorRatioList, clt.cluster_centers_):
      colorList.append(color.astype(np.int64))
      # print(color, colorRatio)


    avg_color = [0, 0, 0]
    for (color, colorRatio) in zip(colorList, colorRatioList):
      if(color[0] != 255 & color[1] != 255 & color[2] != 255):
        avg_color[0] += color[0] * colorRatio * 100
        avg_color[1] += color[1] * colorRatio * 100
        avg_color[2] += color[2] * colorRatio * 100

    avg_color[0] /= 100
    avg_color[1] /= 100
    avg_color[2] /= 100
    
    return avg_color
```

```python
image = cv2.imread(img_url)

# 채널을 BGR -> RGB로 변경
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = image.reshape((image.shape[0] * image.shape[1], 3)) # height, width 통합

count_color = 5
clt = KMeans(n_clusters = count_color)
clt.fit(image)


def centroid_histogram(clt):
    # grab the number of different clusters and create a histogram
    # based on the number of pixels assigned to each cluster
    numLabels = np.arange(0, len(np.unique(clt.labels_)) + 1)
    (hist, _) = np.histogram(clt.labels_, bins=numLabels)

    # normalize the histogram, such that it sums to one
    hist = hist.astype("float")
    hist /= hist.sum()

    # return the histogram
    return hist


hist = centroid_histogram(clt, )
print(hist)
print(type(hist))
[ 0.68881873  0.09307065  0.14797794  0.04675512  0.02337756]

colorList = []
colorRatioList = list(hist)
for (colorRatio, color) in zip(colorRatioList, clt.cluster_centers_):
  colorList.append(color.astype(np.int64))
  # print(color, colorRatio)


avg_color = [0, 0, 0]
for (color, colorRatio) in zip(colorList, colorRatioList):
  if(color[0] != 255 & color[1] != 255 & color[2] != 255):
    avg_color[0] += color[0] * colorRatio * 100
    avg_color[1] += color[1] * colorRatio * 100
    avg_color[2] += color[2] * colorRatio * 100
    
avg_color[0] /= 100
avg_color[1] /= 100
avg_color[2] /= 100

print("평균색", avg_color)

# color[0] /= 100
# color[1] /= 100
# color[2] /= 100

# print(color)


color_hex_list = []

color_hex_list.append(hex(color[0]))
color_hex_list.append(hex(color[1]))
color_hex_list.append(hex(color[2]))

color_hex = color_hex_list[0][-2:] + color_hex_list[1][-2:] + color_hex_list[2][-2:]

# print(color)
# print(color_hex_list)
print(color_hex)
#9C7463
#5D473E 93 71 62
#BC9984 188 153 132
#8C7061 140 112 97


def plot_colors(hist, centroids):
    # initialize the bar chart representing the relative frequency
    # of each of the colors
    bar = np.zeros((50, 300, 3), dtype="uint8")
    startX = 0

    # loop over the percentage of each cluster and the color of
    # each cluster
    for (percent, color) in zip(hist, centroids):
        # plot the relative percentage of each cluster
        endX = startX + (percent * 300)
        cv2.rectangle(bar, (int(startX), 0), (int(endX), 50),
                      color.astype("uint8").tolist(), -1)
        startX = endX

    # return the bar chart
    return bar

bar = plot_colors(hist, clt.cluster_centers_)


show our color bart
plt.figure()
plt.axis("off")
plt.imshow(bar)
plt.show()
```
